# Data Aggregation
* Aggrregation refer to any transformation that refer returning a SCALAR value from arrays
* We so far applied varioues sums, means and etc
* It is the next step after grouping the instances

```python
import numpy as np
import pandas as pd
```

```python
frame = pd.DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'],
                      'key2': ['one', 'two', 'one', 'two', 'one'],
                      'data1': np.random.randn(5),
                      'data2': np.random.randn(5)})
```

```python
frame
```

```python
grouped = frame.groupby('key1')
grouped['data1'].quantile(0.9)
```

```python
grouped.describe()
```

```python
frame
```

```python
grouped = frame.groupby('key1')
grouped.agg('sum')
```

```python
#alternatively
grouped.sum()
```

```python
functions = ['count', 'mean', 'max']
grouped.agg(functions)
```

```python
grouped.agg({'data1': ['min', 'max', 'mean'],
             'data2': ['sum', 'mean']})
```
