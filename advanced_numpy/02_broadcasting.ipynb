# Broadcasting
* Broadcasting is perfromed over missing dimension
* Draw diagrams to imagien what you are dling

```python
import numpy as np
```

## 1. Scalar

```python
array = np.arange(5)
array
```

```python
array * 5
```

## 2. Vector

```python
array = np.random.randn(4, 3)
array
```

```python
np.mean(array, axis=0)
```

```python
# (4,3) - (1,3) = (4,3)
demeaned = array - np.mean(array, axis=0)
demeaned
```

```python
np.mean(demeaned, axis=0)
```

```python
array
```

```python
row_means = np.mean(array, axis=1)
row_means.shape
```

```python
row_means.reshape((4, 1))
```

```python
demeaned = array - row_means.reshape((4, 1))
np.mean(demeaned, axis=1)
```

## 3. Broadcasting Over Other Axes
* Broadcasting dimesnion shoulb be exactly the same dimension as broacasted one: 
* (4,3) => (4, ) x
* (4,3) => (4,1)vv

```python
#np.newaxis inserting the new axis
# the alternative is reshape() => we are comfortable with it

array = np.zeros((4, 4))
array_3d = array[:, np.newaxis, :]
array_3d.shape
```
