# Plotting with pandas and seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Line Plots

### 1.1 Series

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
#by default plot() plots line plot
s.plot()
```

### 1.2 DataFrame

```python
frame = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                     index=np.arange(0, 100, 10),
                     columns=list('ABCD'))
```

```python
#each line = column
frame.plot()
```

```python
frame.plot.line()
```

## 2. Bar Plots
* plot.bar() and barh() plots bars

### 2.1 Series

```python
fig, axes = plt.subplots(2, 1)

data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

### 2.2 DataFrame

```python
frame = pd.DataFrame(np.random.rand(6, 4),
                     index=['one', 'two', 'three', 'four', 'five', 'six'],
                     columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
frame
```

```python
frame.plot.bar()
```

```python
#stakced bar plot => ustiga qoyish
frame.plot.barh(stacked=True, alpha=0.5)
```

### 2.3 Seaborn

```python
sns.barplot(data=frame)
```

## 3. Histograms and Density Plots
* Histograms = discrete plots
* Density = continuous plots
* histograms are the same as bar plots, except: we define # of bins and frequency
* data.plot.hist(bins=n), y axis= frequency
* sns.displot() for histograms and density plots

```python
comp_1 = np.random.normal(0, 1, size=200)
comp_2 = np.random.normal(10, 2, size=200)

values = pd.Series(np.concatenate([comp_1, comp_2]))

sns.displot(values, bins=100, color='k')
```
