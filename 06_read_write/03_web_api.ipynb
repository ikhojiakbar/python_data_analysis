# Interacting with Web APIs

```python
import numpy as np
import pandas as pd
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'

response = requests.get(url)
response
```

```python
#parsing json data to native python dictionary
data = response.json()
data[0]['title']
```

```python
issues = pd.DataFrame(data, columns=['number', 'title', 'labels', 'state'])
issues
```
